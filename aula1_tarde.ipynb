```python
def soma_sibsp_parch(linha):
    return linha['SibSp'] + linha["Parch"]
```

```python
#apply permite aplicar regras em cada linha
nova_coluna = titanic.apply(soma_sibsp_parch, axis=1)
```

```python
nova_coluna.head()
```

```python
type(nova_coluna)
```

```python
titanic["Relatives"] = nova_coluna
```
