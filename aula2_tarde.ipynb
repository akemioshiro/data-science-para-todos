```python
def substitui_nome(linha_do_dataframe):
    nome_mulher = linha_do_dataframe["NomeMulher"]
    nome = linha_do_dataframe["Name"]
    if pandas.isnull(nome_mulher):
        return nome
    else:
        return nome_mulher
    
nova_coluna_name = titanic.apply(substitui_nome, axis=1)
nova_coluna_name.head()
```

```python
titanic["Name"] = nova_coluna_name
titanic.head()
```

```python
tabela_frequencia_cabine_sobrenome = pandas.crosstab(titanic["Cabin"], titanic["Sobrenome"])
tabela_frequencia_cabine_sobrenome
```

```python
pessoas_por_cabine = tabela_frequencia_cabine_sobrenome.sum(axis=1)
pessoas_por_cabine.sort_values(ascending=False).head(10)
```

```python
pessoas_por_familia = tabela_frequencia_cabine_sobrenome.sum(axis=0)
pessoas_por_familia.sort_values(ascending=False).head()
```

```python
pandas.to_datetime(alunos["Data"], format="%d/%m/%Y")
```

```python
def tratar_datas(data):
    return pandas.to_datetime(data, format="%d/%m/%Y")

novas_datas=alunos["Data"].apply(tratar_datas)
novas_datas = alunos["Data"].apply(pandas.to_datetime, format="%d/%m/%Y")
```

```python
alunos["Data"] = novas_datas
alunos.head()
```

```python
def idade(linha):
    data = linha["Data"]
    return datetime.datetime.now().year - data.year

alunos.apply(idade, axis=1)
```

```python
from dateutil.relativedelta import relativedelta

def idade(linha):
    data = linha["Data"]
    return relativedelta(datetime.datetime.now(), data).years

alunos.apply(idade, axis=1)
```
