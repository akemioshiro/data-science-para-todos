```python
def contador_de_caras():
    resultado_das_jogadas = np.random.randint(0,2,50)
    soma = sum(resultado_das_jogadas)
    return soma
```

```python
contador_de_caras()
```

```python
def simulador_jogadas_de_moeda(qtd_de_experiencias):
    qtd_caras = list()
    #repete de acordo com a quantidade de vezes que passamos
    for repeticao in range(qtd_de_experiencias):
        caras = contador_de_caras()
        # pega o resultado da soma e guarda na lista
        qtd_caras.append(caras)
    return qtd_caras
```

```python
def assimetria(coluna):
    moda = coluna.mode()[0]
    media = coluna.mean()
    mediana = coluna.median()
    if(media < media):
        return("negativo")
    elif(media >mediana >moda):
        return ("positivo")
    else:
        return("simetrico")
```

```python
assimetria(titanic.Age)      
```

```python
from scipy.stats import skew # considera todas as linhas do dataframe

skew(titanic.Age.dropna())

def assimetria_stats(coluna):
    skewness = skew(coluna)
    if skewness > 0:
        return "positivo"
    elif skewness < 0:
        return "negativo"
    else:
        return "simetrico"
        
```

```python
def variancia(lista):
    n = len(lista)
    mu = np.mean(lista)
    numerador = 0
    for observacao in lista:
        numerador = numerador + (observacao - mu)**2
    return numerador/n
variancia([10,9,11,12,8])
```

```python
fare_survived_1 = titanic[titanic.Survived ==1].Fare.dropna()
fare_survived_0 = titanic[titanic.Survived ==0].Fare.dropna()
```

```python
plt.figure(figsize=(12,8))
snr.distplot(fare_survived_1, hist=False, label="Survived = 1")
snr.distplot(fare_survived_0, hist=False, label="Survived = 0")
```

```python
ks,p = stats.ks_2samp(fare_survived_0, fare_survived_1).statistic, stats.ks_2samp(fare_survived_0, fare_survived_1).pvalue
```

```python
print("K-S = ", round(ks,4), "p-valor = ", round(p,4))
```

```python
fig,ax = plt.subplots(figsize=(15,8))

#plot the s1 cumulative histogram
n,bins, patches = ax.hist(fare_survived_1[fare_survived_1 < fare_survived_0.max()],500, normed=1, histtype="step", cumulative=True, label="Survived = 1")

n,bins,patches = ax.hist(fare_survived_0, 500, normed=1, histtype="step", cumulative=True, label="Survived = 0")

ax.grid(True)
ax.legend(loc="best")
ax.set_title("Cumulative step histograms")
ax.set_xlabel("Age")
ax.set_ylabel("Likelihood of occurrence")

plt.show()
```
